```python
import pandas as pd

doc = []
with open(r'C:\Users\Himanshu Poddar\Desktop\Coursera courses\Applied text mining with python\dates.txt') as file:
    for line in file:
        doc.append(line)

ori_df = pd.Series(doc)
ori_df.head(10)
```

```python
#Applying Regex with pandas
ori_df = ori_df.str.extract('(\d{1,2}[-/]\d{1,2}[/-]\d{2,4})|(\d{1,2}/\d{4})|((?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[ -.]*\d{2}[thsdn, .-]*\d{4})|((?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[,. ]*\d{4})|(\d{4})')
```

```python
ori_df.iloc[271,4] = ""
ori_df.iloc[271,3] = "August 2008"
```

```python
df = ori_df
df
```

```python
df = df.fillna('')
print(df.head(5))
print(df.tail(5))
```

```python
df
```

```python
for i in range(0,500):
    df.iloc[i][0] = df.iloc[i][0].replace('-','/')
```

```python
# replace 2 digit year with 4 digit year
df[0] = df[0].str.replace(r'/(\d{2})$', r'/19\1')
```

```python
df[0]
```

```python
list(df[2])
```

```python
#converting month name to 3 letter
repl = lambda m: m.group(0)[:3]
df[2] = df[2].str.replace(r'([A-Za-z]+)', repl)
list(df[2])
```

```python
#cleaning the date, replace . and , by space
df[2] = df[2].str.replace('[.,]',' ')
#replace more than one space by /
df[2] = df[2].str.replace('[\s]+','/')
list(df[2])
```

```python
list(df[3])
```

```python
#For data having only mm/yy , assuming it to be the first of that month and then replacing
df[3] = df[3].str.replace(r'(^[a-zA-Z])',r'01 \1')
df[3] = df[3].str.replace(r'([A-Za-z]+)', repl)
```

```python
#cleaning the date, replace . and , by space
df[3] = df[3].str.replace('[.,]',' ')
#replace more than one space by /
df[3] = df[3].str.replace('[\s]+','/')
list(df[3])
```
